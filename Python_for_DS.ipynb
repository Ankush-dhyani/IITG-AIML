How to detect the missing values in your dataset?

```python
uber_df2.isnull().sum()
```

We could infer from the above that null values are present only in the "Purpose" column and that there are 502 null values in the "Purpose" column.

how to impute these null values?

- if the null values are present in a categorical column, impute them with the most frequently occurring values(i.e) with the mode
- if the null values are present in a continuous numerical value column, impute them with either the mean or median values

```python
 uber_df2["PURPOSE*"].value_counts()
```

```python
uber_df2["PURPOSE*"].fillna("Meeting") # fill the null values
```

```python
uber_df2.isnull().sum()
```

if you want the above imputation to reflect in the original dataset , use either of the two commands shown below:

```python
uber_df2["PURPOSE*"] = uber_df2["PURPOSE*"].fillna("Meeting") #assign the value
```

OR

```python
uber_df2["PURPOSE*"].fillna("Meeting", inplace = True) #set the inplace parameter value to True
```

```python
uber_df2.isnull().sum()
```

So , we could confirm from the above code that there are no null values in the dataset
